```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Euclidean Distance

import numpy as np
from scipy.spatial import distance

pointA = np.array([2, 4, 6])
pointB = np.array([5, 1, 9])

euclidean_dist = distance.euclidean(pointA, pointB)

print("Euclidean Distance:", euclidean_dist)

similarity = 1 / (1 + euclidean_dist)

print("Similarity:", similarity)
```

```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Manhattan Distance

import numpy as np
from scipy.spatial import distance

pointA = np.array([2, 4, 6])
pointB = np.array([5, 1, 9])

dist = distance.cityblock(pointA, pointB)

print("Manhattan Distance:", dist)

similarity = 1 / (1 + dist)

print("Similarity:", similarity)
```

```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Minkowski Distance

import numpy as np
from scipy.spatial import distance

pointA = np.array([2, 4, 6])
pointB = np.array([5, 1, 9])

dist = distance.minkowski(pointA, pointB, p=3)

print("Minkowski Distance:", dist)

similarity = 1 / (1 + dist)

print("Similarity:", similarity)
```

```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Hamming Distance

def hamming_distance(str1, str2):

    if len(str1) != len(str2):
        raise ValueError(
            "Strings must be of equal length"
        )

    return sum(
        ch1 != ch2
        for ch1, ch2 in zip(str1, str2)
    )

s1 = "karolin"
s2 = "kathrin"

print(
    "Hamming Distance:",
    hamming_distance(s1, s2)
)
```

```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Jaccard Index

def jaccard_index(str1, str2):

    set1 = set(str1.split())
    set2 = set(str2.split())

    intersection = set1.intersection(set2)
    union = set1.union(set2)

    return len(intersection) / len(union)

s1 = "data science is fun"
s2 = "science makes data useful"

print(
    "Jaccard Index:",
    jaccard_index(s1, s2)
)
```

```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Cosine Similarity

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 = "data science is fun"
s2 = "science makes data useful"

vectorizer = CountVectorizer().fit([s1, s2])

vectors = vectorizer.transform([s1, s2])

cos_sim = cosine_similarity(
    vectors[0],
    vectors[1]
)[0][0]

print("Cosine Similarity:", cos_sim)
```

```python
# Week 3
# Experiment: Similarity and Dissimilarity Measures
# Program: Longest Common Subsequence

def lcs_length(X, Y):

    m = len(X)
    n = len(Y)

    dp = [[0] * (n + 1)
          for _ in range(m + 1)]

    for i in range(m):
        for j in range(n):

            if X[i] == Y[j]:
                dp[i + 1][j + 1] = (
                    dp[i][j] + 1
                )
            else:
                dp[i + 1][j + 1] = max(
                    dp[i][j + 1],
                    dp[i + 1][j]
                )

    return dp[m][n]

print(
    lcs_length(
        "ABCDEF",
        "AEBDF"
    )
)
```
