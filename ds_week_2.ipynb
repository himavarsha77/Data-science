```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Mean and Mode Imputation

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

print(df)
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Forward Fill

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)

print(df_ffill)
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Backward Fill

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_bfill = df.copy()
df_bfill.fillna(method='bfill', inplace=True)

print(df_bfill)
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Deletion of Missing Values

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_drop_rows = df.dropna()

print("After Dropping Rows:")
print(df_drop_rows)

df_drop_cols = df.dropna(axis=1)

print("\nAfter Dropping Columns:")
print(df_drop_cols)
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Exact Match Duplicate Removal

import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original Data")
print(df)

df_exact = df.drop_duplicates()

print("\nAfter Removing Duplicates")
print(df_exact)
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Subset Based Duplicate Removal

import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print(df.drop_duplicates(subset=['ID']))
print(df.drop_duplicates(subset=['Name']))
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Date Standardization

import pandas as pd

df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce'
).dt.strftime('%Y-%m-%d')

print(df)
```

```python
# Week 2
# Experiment: Data Cleaning Techniques
# Program: Case Normalization

import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

df['Name_lower'] = df['Name'].str.lower()
df['Name_upper'] = df['Name'].str.upper()

print(df)
```
