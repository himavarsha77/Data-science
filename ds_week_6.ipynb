```python
# Week 6
# Experiment: Exploratory Data Analysis
# Program: Measures of Central Tendency (Mean, Median, Mode)

import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 Rows:")
print(df.head())

print("\nMean Values:")
print(df.mean())

print("\nMedian Values:")
print(df.median())

print("\nMode Values:")
print(df.mode().iloc[0])
```

```python
# Week 6
# Experiment: Exploratory Data Analysis
# Program: Measures of Dispersion (Range, Variance, Standard Deviation, IQR)

import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 Rows:")
print(df.head())

# Range
range_values = df.max() - df.min()
print("\nRange:")
print(range_values)

# Variance
print("\nVariance:")
print(df.var())

# Standard Deviation
print("\nStandard Deviation:")
print(df.std())

# IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print("\nInterquartile Range (IQR):")
print(IQR)
```

```python
# Week 6
# Experiment: Exploratory Data Analysis
# Program: Skewness and Kurtosis Analysis

import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 Rows:")
print(df.head())

print("\nSkewness:")
print(df.skew())

print("\nKurtosis:")
print(df.kurt())
```
